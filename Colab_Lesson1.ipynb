```python
import numpy as np
import matplotlib.pyplot as plt

# --- Input Parameters ---
N0 = 1000         # Initial amount (e.g., number of atoms)
half_life = 8.0   # Half-life in hours
t_max = 48        # Max time in hours
num_points = 100  # Resolution

# --- Calculations ---
lambda_decay = np.log(2) / half_life
t = np.linspace(0, t_max, num_points)
N = N0 * np.exp(-lambda_decay * t)

# --- Plot ---
plt.figure(figsize=(8, 5))
plt.plot(t, N, label='Radioactive Decay', color='red', linestyle = '--') # color and linestyle
plt.xlabel('Time (hours)')
plt.ylabel('Remaining Amount')
plt.title('Radioactive Decay Curve')
plt.grid(True)
plt.legend()
plt.show()

```
